## 2.2. Finding Associations

Compute Frequent Itemsets using Apriori

```python
from mlxtend.frequent_patterns import apriori
```

```python
#Compute itemsets with min_support = 30% with item names and with length
frequent_itemsets = apriori(binary_database, min_support=0.3, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

It would be great if the results were sorted. Also lets look for length > 1 

```python
pd.options.display.max_rows = 20
frequent_itemsets[(frequent_itemsets['length']>1)].sort_values(by=['support'], ascending=False)
```

Generate Association Rules from Frequent Itemsets

Essencially, the known that:

> **High Suppot:**     Should apply to large amount of cases

> **High Confidence:** Should be Correct often

> **High Lift:**       Indicates that this rules aren't just a coicidence

```python
from mlxtend.frequent_patterns import association_rules
pd.options.display.max_rows = 10

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.001,support_only=False)
a = rules.sort_values(by=['support'], ascending=False)
a[(a['conviction'] < np.inf)]
```

Now... The most commumn type of mental disorder is, by far, Mood Disorder (405 entries in 556 possible ones). For now, i'm focusing in it.

```python
mood_disorder = rules[(rules['support']>0.6) & (rules['lift']>1)& (rules['confidence']>0.6)]
#I am converting this to str to search for the disorder I want
mood_disorder['consequents'] = mood_disorder['consequents'].astype('str')
b = mood_disorder[(mood_disorder['consequents'].str.contains("q49_Mood Disorder") )]

pd.options.display.max_rows = 20

b.sort_values(by=['confidence'], ascending=False).head(10)
```

So, when speaking about mood disorder, the entries are related with four questions: 25, 47, 51, 53 which are the sames as before


I wonder if, considering other possible disease, the rules stand. I will allow a smaller support, and restrict for search into the second largest present disease : "Anxiety Disorder" (338 in 556). For the other disorders, their count is really small, and we will ignore then.

```python
other_disorder = rules[(rules['support']>0.5) & (rules['lift']>1)& (rules['confidence']>0.6)]

other_disorder['consequents'] = other_disorder['consequents'].astype('str')
other_disorder['consequents'].dtype
c = other_disorder[(other_disorder['consequents'].str.contains("Anxiety Disorder"))]

c.sort_values(by=['lift'], ascending=False).head(10)
```

So, I'm observing the same kind of pattern: the same questions.
