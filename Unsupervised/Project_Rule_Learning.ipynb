## 2.2. Finding Associations

Compute Frequent Itemsets using Apriori

```python
from mlxtend.frequent_patterns import apriori
```

```python
#Compute itemsets with min_support = 60% with item names and with length
frequent_itemsets = apriori(binary_database, min_support=0.3, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

Generate Association Rules from Frequent Itemsets

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules with confidence >= 90%
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.001,support_only=False)
a = rules.sort_values(by=['support'], ascending=True)
#a[(a['conviction'] < np.inf)]
```

Essencially, the have that:

> **High Suppot:**     Should apply to large amount of cases

> **High Confidence:** Should be Correct often

> **High Lift:**       Indicates that this rules aren't just a coicidence

Now... The most commumn type of mental disorder is, by far, Mood Disorder (405 entries in 556 possible ones). For now, i'm focusing in it.

```python
mood_disorder = rules[(rules['support']>0.6) & (rules['lift']>1)& (rules['confidence']>0.6)]
mood_disorder = mood_disorder.sort_values(by=['lift'], ascending=False)
#mood_disorder
```
